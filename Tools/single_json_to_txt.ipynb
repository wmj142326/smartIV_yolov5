```python
import pandas as pd
import numpy as np
import os
import cv2
import numpy as np
import shutil
```

```python
'''
下面有两目录,labels和images,分别以json和jpg结尾
最终输出在txt文件夹中所有的labels的json转txt文件
'''
```

```python
label_path = './labels'
label_path_list = []
image_path_list = []
for home, dirs, files in os.walk(label_path):
    for file in files:
        label_path = os.path.join(home,file)
        label_path_list.append(label_path)
        img_path = label_path.replace('labels',"images").replace("json","jpg")
        image_path_list.append(img_path)
```

```python
txt_dir='./txt'
if os.path.exists(txt_dir):
    print('true')
    # os.rmdir(txt_dir)
    shutil.rmtree(txt_dir) # 删除再建立
    os.makedirs(txt_dir)
else:
    os.makedirs(txt_dir)
```

```python
def split_col(data, columns):
    """拆分成列

    :param data: 原始数据
    :param columns: 拆分的列名
    :type data: pandas.core.frame.DataFrame
    :type columns: list
    """
    for c in columns:
        new_col = data.pop(c)
        max_len = max(list(map(len, new_col.values)))  # 最大长度
        new_col = new_col.apply(lambda x: x + [None]*(max_len - len(x)))  # 补空值，None可换成np.nan
        new_col = np.array(new_col.tolist(),dtype=np.float32).T  # 转置
        for i, j in enumerate(new_col):
            data[c + str(i)] = j
```

```python
for i,label_path in enumerate(label_path_list):
    df = pd.read_json(label_path)
    df=df[['category_name','bbox']]
    # print(df)

    # map_category = pd.Series({"轿车":0,"成年人":1,"suv":2,"电动/摩托车":3,"面包车":4,"自行车":5,"大客车":6,"大货车":7,"电动/摩托三轮车":8,"小货车":9,"微型车":10,"专业作业车":11,"儿童":12,"其他":13,"人力三轮车":14,"其他非机动车,模糊或变形严重车辆":15})
    map_category = pd.Series({"轿车":0,"suv":1,"SUV":1,"电动/摩托车":2,"电动车/摩托车":2,"面包车":3,"自行车":4,"大客车":5,"大货车":6,"电动/摩托三轮车":7,"电动三轮车/摩托三轮车":7,"小货车":8,"微型车":9,"专业作业车":10,"人力三轮车":11})
    df['category_name'] = df['category_name'].map(map_category)
    # print(df['category_name'])
    df.dropna(inplace=True)
    df['category_name'] = df['category_name'].astype('int')
    # print(df['category_name'])
    
    
    label_img_path = label_path.replace('labels','images').replace('.json','.jpg') 
    img_shape = cv2.imread(label_img_path).shape
    img_h = img_shape[0]
    img_w = img_shape[1]

    split_col(df, columns=['bbox'])
    df['bbox0'] = df['bbox0']/img_w
    df['bbox1'] = df['bbox1']/img_h
    df['bbox2'] = df['bbox2']/img_w
    df['bbox3'] = df['bbox3']/img_h
    # print(df)

    df = df.dropna(axis=0, subset=['category_name'])
    
    txt_name = label_path.replace('labels','txt').replace('.json','.txt') 
    df.to_csv(txt_name,header=None,index=None,sep='\t')
    print("第{}张已完成，共{}张".format(i+1,len(label_path_list)))

    
```
